```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/content/entrepreneurship_decision_dataset.csv"  # Update path as needed
df = pd.read_csv(file_path)

# Group by Industry_Type and calculate average Risk_Tolerance and Investment_Availability
industry_stats = df.groupby("Industry_Type")[["Risk_Tolerance", "Investment_Availability"]].mean().reset_index()

# Set up the visualization
plt.figure(figsize=(14, 6))

# Plot average risk tolerance
plt.subplot(1, 2, 1)
sns.barplot(x="Risk_Tolerance", y="Industry_Type", data=industry_stats, palette="Blues_d")
plt.title("Average Risk Tolerance by Industry")
plt.xlabel("Average Risk Tolerance")
plt.ylabel("Industry Type")

# Plot average investment availability
plt.subplot(1, 2, 2)
sns.barplot(x="Investment_Availability", y="Industry_Type", data=industry_stats, palette="Greens_d")
plt.title("Average Investment Availability by Industry")
plt.xlabel("Average Investment Availability")
plt.ylabel("")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/content/entrepreneurship_decision_dataset.csv"
df = pd.read_csv(file_path)

# Set style for aesthetics
sns.set(style="whitegrid")

# Heatmap for correlation
plt.figure(figsize=(6, 4))
correlation_matrix = df[['Growth_Rate', 'Risk_Tolerance']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f")
plt.title("Heatmap: Correlation between Growth Rate and Risk Tolerance")
plt.tight_layout()
plt.show()

# Create string-based bins to avoid dtype issues
df['Growth_Rate_Bin'] = pd.qcut(df['Growth_Rate'], q=10, duplicates='drop')
df['Growth_Rate_Bin_Label'] = df['Growth_Rate_Bin'].astype(str)

# Line plot using bin labels
plt.figure(figsize=(10, 5))
sns.lineplot(
    x='Growth_Rate_Bin_Label',
    y='Risk_Tolerance',
    data=df,
    estimator='mean',
    errorbar='sd',
    marker='o',
    sort=False,
    linewidth=2
)
plt.xticks(rotation=45)
plt.title("Mean Risk Tolerance Across Growth Rate Bins")
plt.ylabel("Average Risk Tolerance")
plt.xlabel("Growth Rate (Binned)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/entrepreneurship_decision_dataset.csv')

# Compute the correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Extract correlations with Risk_Tolerance
risk_corr = corr_matrix["Risk_Tolerance"].drop("Risk_Tolerance").sort_values(key=abs, ascending=False)

# Plot the correlations with Risk_Tolerance
plt.figure(figsize=(8, 5))
sns.barplot(x=risk_corr.values, y=risk_corr.index, palette="coolwarm")
plt.title("Correlation of Risk Tolerance with Other Factors")
plt.xlabel("Correlation Coefficient")
plt.axvline(0, color='black', linestyle='--')
plt.tight_layout()
plt.show()
```
